```python
# text preprocessing befor text similarity ML using NLP 

import spacy 

nlp = spacy.load("en_core_web_sm")

resume_doc = nlp(resume)

def preprocessing(text):

    doc = nlp(text)

    tokens = [token.lemma_.strip().lower() for token in doc if not token.is_stop and not token.is_punct]

    preprocessed_text = " ".join(tokens)

    return preprocessed_text

```

```python
search_term = {"Data Scientist"} # change this later 
search_location = "London"

params = {
    "q": search_term,
    "location": search_location,
    "engine": "google_jobs",
    "hl": 'en',
    "gl": "uk",
    "api_key": api_key,
    "chips": "date_posted:today"
}

search = GoogleSearch(params)

```

```python
results = search.get_dict()
jobs_df = results["jobs_results"]
jobs_df = pd.DataFrame(jobs_df)

columns_to_keep = ['title', 'company_name', 'location', 'apply_options', 'description']
jobs_df = jobs_df[columns_to_keep]

links = []
for i in range(len(jobs_df)):
    links.append(jobs_df['apply_options'][i][0]['link'])

jobs_df['links'] = links
jobs_df.drop('apply_options', axis=1, inplace=True)
```

```python
# Cleaning the jobs descriptions using NLP 

resume = preprocessing(resume)
jobs_df["description"] = jobs_df["description"].apply(preprocessing)
```

```python
import numpy as np
from sentence_transformers import SentenceTransformer 
```

```python
import warnings
warnings.filterwarnings('ignore')

def distilBERT_compare(CV, job_desc):
    model = SentenceTransformer("distilbert-base-nli-mean-tokens")

    encoding1 = model.encode(CV)
    encoding2 = model.encode(job_desc)

    similarty = np.dot(encoding1, encoding2) / (np.linalg.norm(encoding1) * np.linalg.norm(encoding2))

    return similarty

descriptions = jobs_df["description"]

BERT_percentage = []
for i in range(len(jobs_df)):
    BERT_percentage.append(distilBERT_compare(resume, descriptions.iloc[i])*100)

jobs_df['distil BERT Match %'] = BERT_percentage
jobs_df['distil BERT Match %'] = jobs_df['distil BERT Match %'].round(2)

# sorting for highest % matches at the top
jobs_df.sort_values(by="distil BERT Match %", ascending=False, inplace=True)
# resetting index so they stay in ascending order
jobs_df.reset_index(drop=True, inplace=True)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

descriptions = jobs_df["description"]

# comparison function using cosine similarity 
def compare(CV, job_desc):
    compare = [CV, job_desc]
    cVect = CountVectorizer()
    cMatrix = cVect.fit_transform(compare)

    matPerecent = cosine_similarity(cMatrix)[0][1] * 100 
    matPerecent = round(matPerecent, 2)

    return matPerecent

match_percentage =[]
for i in range(len(jobs_df)):
    match_percentage.append(compare(resume, descriptions.iloc[i]))

jobs_df["Cosine Match %"] = match_percentage

#sorting for highest % matches at the top
jobs_df.sort_values(by="Cosine Match %", ascending=False, inplace=True)
# resetting index so they stay in ascending order
jobs_df.reset_index(drop=True, inplace=True)

jobs_df
```

```python
# Filtering for entry level jobs 
search_words = ["Senior", "Head", "Chief", "Lead", "Manager"]
search_words = [word.lower() for word in search_words]

jobs_df['title'] = jobs_df["title"].str.lower()

filter = jobs_df["title"].apply(lambda x: any(term in x for term in search_words)) # finds all the rows which contain this
# for just one word we can just use .contains()

jobs_df = jobs_df[~filter] # this symbol ~ is the NOT operator 

jobs_df.reset_index(drop=True, inplace=True)
jobs_df
```
